### 3.Handle missing values (identify, fill/drop). 

```python
df.isnull().sum()
```

```python
df_cleaned=df.copy()
```

```python
df_cleaned.drop(columns=["videos","variations"],inplace=True)
```

```python
df_cleaned["discount"]=df["discount"].fillna(0)
df_cleaned["what_customers_said"]=df["what_customers_said"].fillna("No review available")
df_cleaned["seller_name"]=df["seller_name"].fillna("Unknown")
df_cleaned["seller_information"]=df["seller_information"].fillna("Not available")
```

```python
df_cleaned.isnull().sum()
```

```python
#Highly rated watches
df_cleaned[(df_cleaned["rating"]>4.5) & (df_cleaned["category"]=="watches")]
```

```python
#Converting string prices to float prices 
df_cleaned["final_price"] = (
    df_cleaned["final_price"].str.replace("₹", "", regex=False).str.replace(",", "", regex=False)
    .str.replace('"', "", regex=False).str.strip().astype(float)
)
```

```python
df_cleaned["final_price"]
```

```python
#Accessing rows with category as Jeans and price <1000
df_cleaned[(df_cleaned['category'] == 'jeans') & (df_cleaned['final_price'] < 1000)]
```

```python
#Accessing high rated products and price<500
df_cleaned[(df_cleaned["rating"]>4.5) & (df_cleaned['final_price'] < 500)]
```
